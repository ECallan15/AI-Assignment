```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("card_transdata.csv")

# Display the first 5 rows of the dataset
print("Preview of the dataset:")
df.head()
```

```python
# Display the shape of the dataset
print("Dataset shape (rows, columns):", df.shape)
```

```python
# Display dataset information
print("Dataset Information:")
df.info()
```

```python
# Display statistical summary of the dataset
print("Statistical Summary:")
df.describe()
```

```python
# Check for missing values in the dataset
print("Missing Values in Each Column:")
print(df.isnull().sum())
```

```python
# Drop rows with any null values and create a clean copy
df = df.dropna().copy()

# Display the new shape after dropping missing values
print(df.shape)
```

```python
# Display the first 5 rows of the cleaned dataset
df.head()
```

```python
# Updated: Bar chart of Fraud Class Distribution (no FutureWarning)
plt.figure(figsize=(8, 5))
sns.countplot(x='fraud', hue='fraud', data=df, palette='coolwarm', legend=False)

# Add title and labels
plt.title("Fraud Class Distribution", fontsize=14)
plt.xlabel("Fraud", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Display the chart
plt.show()
```

```python
# Bar chart of Used Pin Number Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='used_pin_number', hue='used_pin_number', data=df, palette='coolwarm', legend=False)

# Add title and labels
plt.title("Used Pin Number Distribution", fontsize=14)
plt.xlabel("Used Pin Number", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Display the chart
plt.show()
```

```python
# Bar chart of Repeat Retailer Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='repeat_retailer', hue='repeat_retailer', data=df, palette='coolwarm', legend=False)

# Add title and labels
plt.title("Repeat Retailer Distribution", fontsize=14)
plt.xlabel("Repeat Retailer", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Display the chart
plt.show()
```

```python
# Bar chart of Online Order Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='online_order', hue='online_order', data=df, palette='coolwarm', legend=False)

# Add title and labels
plt.title("Online Order Distribution", fontsize=14)
plt.xlabel("Online Order", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Display the chart
plt.show()
```

```python
# Histogram of Ratio to Median Purchase Price
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='ratio_to_median_purchase_price', bins=30, kde=True, color='steelblue')

# Add title and labels
plt.title("Distribution of Ratio to Median Purchase Price", fontsize=14)
plt.xlabel("Ratio to Median Purchase Price", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Display the chart
plt.show()
```

```python
# Count plot: Transactions with PIN vs Fraud
plt.figure(figsize=(8, 5))
sns.countplot(x='used_pin_number', hue='fraud', data=df, palette='coolwarm')

# Add title and labels
plt.title("Transactions with PIN vs Fraudulent Transactions", fontsize=14)
plt.xlabel("Used PIN Number", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="Fraud", labels=["Non-Fraudulent", "Fraudulent"])

# Display the chart
plt.show()
```

```python
# Calculate counts of fraud and non-fraud by PIN usage
pin_fraud_counts = pd.crosstab(df['used_pin_number'], df['fraud'])

# Plot stacked bar chart
pin_fraud_counts.plot(kind='bar', stacked=True, figsize=(8, 5), color=['skyblue', 'salmon'])

# Add title and labels
plt.title("Transactions with PIN vs Fraudulent Transactions", fontsize=14)
plt.xlabel("Used PIN Number", fontsize=12)
plt.ylabel("Number of Transactions", fontsize=12)
plt.legend(title="Fraud", labels=["Non-Fraudulent", "Fraudulent"])

# Display the chart
plt.show()
```

```python
# Count plot: Transactions with Chip vs Fraud
plt.figure(figsize=(8, 5))
sns.countplot(x='used_chip', hue='fraud', data=df, palette='coolwarm')

# Add title and labels
plt.title("Transactions with Chip vs Fraudulent Transactions", fontsize=14)
plt.xlabel("Used Chip", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="Fraud", labels=["Non-Fraudulent", "Fraudulent"])

# Display the chart
plt.show()
```

```python
# Calculate counts of fraud and non-fraud by chip usage
chip_fraud_counts = pd.crosstab(df['used_chip'], df['fraud'])

# Plot stacked bar chart
chip_fraud_counts.plot(kind='bar', stacked=True, figsize=(8, 5), color=['skyblue', 'salmon'])

# Add title and labels
plt.title("Transactions with Chip vs Fraudulent Transactions", fontsize=14)
plt.xlabel("Used Chip", fontsize=12)
plt.ylabel("Number of Transactions", fontsize=12)
plt.legend(title="Fraud", labels=["Non-Fraudulent", "Fraudulent"])

# Display the chart
plt.show()
```

```python
# Calculate counts of fraud and non-fraud by online order
online_fraud_counts = pd.crosstab(df['online_order'], df['fraud'])

# Plot stacked bar chart
online_fraud_counts.plot(kind='bar', stacked=True, figsize=(8, 5), color=['skyblue', 'salmon'])

# Add title and labels
plt.title("Transactions with Online Order vs Fraudulent Transactions", fontsize=14)
plt.xlabel("Online Order", fontsize=12)
plt.ylabel("Number of Transactions", fontsize=12)
plt.legend(title="Fraud", labels=["Non-Fraudulent", "Fraudulent"])

# Display the chart
plt.show()
```

```python
# Convert fraud to categorical for proper color mapping
df['fraud_cat'] = df['fraud'].map({0: 'Non-Fraudulent', 1: 'Fraudulent'})

# Scatter plot with correct colors
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='distance_from_home',
    y='ratio_to_median_purchase_price',
    hue='fraud_cat',
    data=df,
    palette={'Non-Fraudulent': 'green', 'Fraudulent': 'red'},
    alpha=0.6
)

# Add title and labels
plt.title("Fraudulent vs Non-Fraudulent Transactions: Distance from Home vs Ratio to Median Purchase Price", fontsize=14)
plt.xlabel("Distance from Home", fontsize=12)
plt.ylabel("Ratio to Median Purchase Price", fontsize=12)
plt.legend(title="Fraud")

# Display the plot
plt.show()
```
